```python
import os
import re
from datetime import date
import pandas as pd

```

```python
df = pd.read_csv(os.path.join('..', 'data', 'interim', 'job_offers_2024_03_18.csv'), sep=';')
df
```

### wyrzuć oferty bez kwoty w PLN

```python
df_copy = df.copy(deep=True)
```

```python
df_copy = df_copy[df_copy['salary.currency'] == 'PLN']
df_copy
```

```python
df_copy.columns = [col.replace('.', '_') for col in df_copy.columns]
```

```python
df_copy
```

### average salary

```python
df_copy['salary_avg'] = df_copy[['salary_low', 'salary_high']].mean(axis=1)
df_copy
```

### unifikacja nazw miast

```python
cities = df_copy['location_city'].unique()
```

```python
cities.sort()
cities
```

```python
df_copy.loc[df_copy['location_city'].isin(['Warszawa', 'Warsaw']), 'location_city'] = 'Warszawa'
```

```python
df_copy.loc[df_copy['location_city'].isin(['Wrocław', 'Wroclaw']), 'location_city'] = 'Wrocław'
df_copy.loc[df_copy['location_city'].isin(['Cracow', 'Kraków']), 'location_city'] = 'Kraków'
```

```python
df_copy
```

```python
df_copy.loc[df_copy['location_country'].isnull(), 'location_country'] = 'B/D'
```

```python
df_copy
```

### dodaj kolumnę is_senior

```python
print(1 if re.search('(?i)(senior)', 'Data senior Engineer') else 0)
```

```python
df_copy['is_senior'] = df_copy['name'].apply(lambda x: 1 if re.search('(?i)(senior)', str(x)) else 0)
```

```python
df_copy
```

```python
date_ = date.today().strftime("%Y_%m_%d")
df_copy.to_csv(os.path.join('..', 'data', 'processed', f'job_offers_{date_}.csv'), sep=';', index=False, encoding='UTF-8')
```
